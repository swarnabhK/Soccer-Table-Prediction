```python
import math
```

```python
def elo(h_rating,a_rating,Home_goals,Away_goals):
    h_expected = 1/(1+math.pow(10,(a_rating-h_rating)/400))
    a_expected = 1/(1+math.pow(10,(h_rating-a_rating)/400))
    if(Home_goals==Away_goals):
        h_rating = h_rating+50*(0.5-h_expected)
        a_rating = a_rating+50*(0.5-a_expected)
    elif(Home_goals>Away_goals):
        h_rating = h_rating+50*(1-h_expected)
        a_rating = a_rating+50*(0-a_expected)
    else:
        h_rating = h_rating+50*(0-h_expected)
        a_rating = a_rating+50*(1-a_expected)
    return h_rating,a_rating
```

```python
w_rating,l_rating = elo(1000,1000,0,0)
w_rating2,l_rating2 = elo(1000,1000,3,0)
w_rating3,l_rating3 = elo(1000,1000,0,3)
print(w_rating,l_rating)
print(w_rating2,l_rating2)
print(w_rating3,l_rating3)
```
